```python
# Level 14
fn = "../input/wire.png"
img = Image.open(fn)
delta = [(1, 0), (0, 1), (-1, 0), (0, -1)]

out = Image.new('RGB', [100, 100])
x, y, p = -1, 0, 0
d = 200
while d/2 > 0:
    for v in delta:
        steps = d // 2
        for s in range(steps):
            x, y = x + v[0], y + v[1]
            out.putpixel((x,y), img.getpixel((p,0)))
            p += 1
        d -= 1
out
# uzi
```

```python
# Level 18
import gzip, difflib

fn = "../input/deltas.gz"
d1, d2 = [], []
with gzip.open(fn) as f:
    for line in f:
        data = line.decode()
        d1.append(data[:53]+"\n")
        d2.append(data[56:])
```

```python
compare = difflib.Differ().compare(d1, d2)

f = open("f.png", 'wb')
f1 = open("f1.png", 'wb')
f2 = open("f2.png", 'wb')

for line in compare:
    try:
        bs = bytes([int(o, 16) for o in line[2:].strip().split(" ")])
        if line[0] == "+":
            f1.write(bs)
        elif line[0] == "-":
            f2.write(bs)
        else:
            f.write(bs)
    except:
        # print(line)
        pass

f.close()
f1.close()
f2.close()
```
